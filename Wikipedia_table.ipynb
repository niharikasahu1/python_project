```python
import pandas as pd
```

```python
# Download the every tables in the page
dfs = pd.read_html("https://en.wikipedia.org/wiki/List_of_weather_records")
```

```python
#Find the total number of tables
len(dfs)
```

##1st table

```python
# View the 1st table
table1 = dfs[1]
table1
```

```python
a = table1.copy()
```

```python
a
```

```python
# Transfering the row data to readable data
table1[["Temperature(C)","Temperature(F)"]] = table1.Temperature.str.split("(",expand=True)
table1[["Town","a"]] = table1["Town/Location"].str.split(",",n=1,expand=True)
table1[["Country","b"]] = table1["Country/Region"].str.split("(",n=1,expand=True)
table1[["date","c"]] = table1["Date"].str.split("[",n=1,expand=True)
```

```python
# Drop the unnecessery columns
table1.drop(["Temperature","Town/Location","Country/Region","a","b","c","Date"],axis=1,inplace=True)
```

```python
table1
```

```python
#
table1["Temperature(C)"] = [x[:-2] for x in table1["Temperature(C)"]]
table1["Temperature(F)"] = [x[:-2] for x in table1["Temperature(F)"]]
```

```python
# Check the datatypes
table1.dtypes
```

```python
# date convert to datetime object
table1["date"] = pd.to_datetime(table1["date"])
```

```python
table1
```

```python
# Reindex the column
table1 = table1.reindex(["date","Town","Country","Temperature(C)","Temperature(F)"],axis=1)
```
